```python
import pandas as pd
import numpy as np
df = pd.read_csv("train.csv", low_memory=False)
df.head()
```

```python
df.drop(["Name", "SSN", "ID", "Customer_ID"], axis=1, inplace=True, errors="ignore")
```

```python
def adjustCreditHistoryAge(text):
    if pd.isna(text):
        return 0
    
    parts = text.split()
    years = int(parts[0])
    
    if len(parts) > 3:
        months = int(parts[3])
    else:
        months = 0
        
    totalMonths = (years * 12) + months
    
    return totalMonths

df['Credit_History_Age'] = df['Credit_History_Age'].apply(adjustCreditHistoryAge)
df.head(20)
```

```python
df.head()
```

```python
def replaceSpecialCharacters(text):
        if "_" not in str(text):
            return text

        eraseChars = str(text).replace("_", "")
        try:
            eraseChars = pd.to_numeric(eraseChars)
        except ValueError as e:
            eraseChars = np.nan
        return np.nan if eraseChars == "nan" or eraseChars == "" else eraseChars
    
df['Age'] = pd.to_numeric(df['Age'].apply(replaceSpecialCharacters))
df['Annual_Income'] = pd.to_numeric(df['Annual_Income'].apply(replaceSpecialCharacters))
df['Changed_Credit_Limit'] = pd.to_numeric(df['Changed_Credit_Limit'].apply(replaceSpecialCharacters))
df['Outstanding_Debt'] = pd.to_numeric(df['Outstanding_Debt'].apply(replaceSpecialCharacters))
df['Num_of_Delayed_Payment'] = pd.to_numeric(df['Num_of_Delayed_Payment'].apply(replaceSpecialCharacters))
df['Amount_invested_monthly'] = pd.to_numeric(df['Amount_invested_monthly'].apply(replaceSpecialCharacters))
df['Monthly_Balance'] =  pd.to_numeric(df['Monthly_Balance'].apply(replaceSpecialCharacters))
df["Num_of_Loan"] =  pd.to_numeric(df["Num_of_Loan"].apply(replaceSpecialCharacters))
df.head(20)
```

```python
def removeAgeOutliers(text):
    try:
        text = pd.to_numeric(text, errors='raise')
        
        if text < 0 or text >= 110:
            return np.nan
        else:
            return text
    except ValueError:
        return np.nan

def removeOtherOutliers(text):
    try:
        text = pd.to_numeric(text, errors='raise')
        
        if text < 0:
            return np.nan
        else:
            return text
    except ValueError:
        return np.nan

df['Age'] = df['Age'].apply(removeAgeOutliers)
df['Num_of_Delayed_Payment'] = df['Num_of_Delayed_Payment'].apply(removeOtherOutliers)
df['Num_of_Loan'] = df['Num_of_Loan'].apply(removeOtherOutliers)
df.head(20)
```

```python
df = df.dropna(subset=['Num_of_Loan'])
df = df.dropna(subset=['Type_of_Loan'])
df.head(20)
```

```python
typesOfLoan=df['Type_of_Loan'].str.split(',\s*and\s*|\s*,\s*')
typesOfLoan = typesOfLoan.explode()
typesOfLoanLabels = typesOfLoan.unique()
```

```python
def removeAnd(text):
    andRemoved = str(text).replace(" and", "")
    if andRemoved == "nan":
        return np.nan
    return andRemoved
```

```python
df['Type_of_Loan'] = df['Type_of_Loan'].apply(removeAnd)

for typesOfLoanLabel in typesOfLoanLabels:
    df['Count_' + typesOfLoanLabel] = 0

for index, row in df.iterrows():
    if pd.notna(row['Type_of_Loan']):
        loanTypes = row['Type_of_Loan'].split(', ')
        for loanType in loanTypes:
            if loanType in typesOfLoanLabels:
                df.at[index, 'Count_' + loanType] += 1
```

```python
df.drop(['Num_of_Loan'], axis=1, inplace=True, errors="ignore")
```

```python
df.drop(["Type_of_Loan"], axis=1, inplace=True, errors="ignore")
```

```python

df.dropna(inplace=True)
df["Age"] = pd.to_numeric(df["Age"])
df["Annual_Income"] = pd.to_numeric(df["Annual_Income"])
df["Num_of_Delayed_Payment"] = pd.to_numeric(df["Num_of_Delayed_Payment"])
df["Changed_Credit_Limit"] = pd.to_numeric(df["Changed_Credit_Limit"])
df["Outstanding_Debt"] = pd.to_numeric(df["Outstanding_Debt"])
df["Amount_invested_monthly"] = pd.to_numeric(df["Amount_invested_monthly"])
df["Monthly_Balance"] = pd.to_numeric(df["Monthly_Balance"])
```

```python
df.head()
```

```python
df.drop(df[df['Occupation'] == '_______'].index, inplace = True)
df.drop(df[df['Payment_of_Min_Amount'] == 'NM'].index, inplace = True)
df.drop(df[df['Credit_Mix'] == '_'].index, inplace = True)
df.drop(df[df['Payment_Behaviour'] == '!@9#%8'].index, inplace = True)
print('Number of Rows with Missing Values: ', df.isnull().any(axis=1).sum())
print('Number of Rows: ', df.any(axis=1).sum())
total_nans = df.isna().sum().sum()

print(f'Total number of NaNs: {total_nans}')
```

```python
df = pd.get_dummies(df, columns=['Occupation', 'Credit_Mix','Payment_of_Min_Amount'], dtype=int)
df.head(20)
```

```python
df['Spent_Amount_Payment_Behaviour'] = df['Payment_Behaviour'].str.extract(r'(\w+)_spent')
df['Value_Amount_Payment_Behaviour'] = df['Payment_Behaviour'].str.extract(r'_spent_(\w+)_value')
df.drop(["Payment_Behaviour"], axis=1, inplace=True, errors="ignore")

df.head(20)
```

```python
spentMapping = {'Low': 0, 'High': 1}
valueMapping = {'Small': 0, 'Medium': 1, 'Large': 2}
monthMapping = {'January': 1,'February': 2,'March': 3,'April': 4,'May': 5,'June': 6,'July': 7,'August': 8,'September': 9,'October': 10,'November': 11,'December': 12}

df['Spent_Amount_Payment_Behaviour'] = df['Spent_Amount_Payment_Behaviour'].map(spentMapping)
df['Value_Amount_Payment_Behaviour'] = df['Value_Amount_Payment_Behaviour'].map(valueMapping)
df['Month'] = df['Month'].map(monthMapping)

df.head(20)
```

```python
creditScoreMapping = {'Poor': 0, 'Standard': 1, 'Good': 2}
df['Credit_Score'] = df['Credit_Score'].map(creditScoreMapping)

df['Credit_Score'].head(20)
```

```python
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import os

originalCounts = df['Credit_Score'].value_counts()

X = df.drop('Credit_Score', axis = 1)
Y = df['Credit_Score']

smote = SMOTE(random_state = 42)

resampledX, resampledY = smote.fit_resample(X, Y)

print(resampledY.value_counts())

resampledCounts = resampledY.value_counts()

countsDf = pd.DataFrame({
    'Before': originalCounts,
    'After': resampledCounts
})

countsDf.plot(kind='bar', figsize=(10, 6))
plt.title('Number of Data Instances Before and After SMOTE')
plt.xlabel('Credit Score')
plt.ylabel('Number of Instances')
plt.xticks(ticks=range(len(countsDf)), labels=['Poor', 'Standard', 'Good'], rotation=0)
plt.show()

featuresToInvestigate = resampledX.columns
correlation_matrix = pd.DataFrame(resampledX, columns=featuresToInvestigate).corr()

plt.figure(figsize=(50, 20))
sns.heatmap(
    correlation_matrix, 
    annot=True, 
    cmap='coolwarm', 
    fmt='.2f', 
    linewidths=0.5,
    cbar_kws={'shrink': 0.5},
    annot_kws={"size": 10}
)
plt.title('Correlation Matrix Heatmap')
downloads_path = os.path.expanduser('~/Downloads/correlation_matrix_heatmap.pdf')

plt.savefig(downloads_path, format='pdf', bbox_inches='tight')
plt.close()


print("Correlation Matrix saved as 'correlation_matrix_heatmap.pdf'")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import ElasticNet

standardScaler = StandardScaler()
scaledResampledX = standardScaler.fit_transform(resampledX)

elasticNet = ElasticNet(alpha=0.15, l1_ratio = 0.1, random_state = 42)
elasticNet.fit(scaledResampledX, resampledY)
coefficients = elasticNet.coef_
featureNames = X.columns

coefDf = pd.DataFrame({'Feature': featureNames, 'Coefficient': coefficients})

coefDf
```

```python
nonZeroFeatures = coefDf[coefDf['Coefficient'] != 0]['Feature']
nonZeroFeatures
```

```python
import matplotlib.pyplot as plt

nonZeroFeaturesDf = coefDf[coefDf['Coefficient'] != 0].copy()

nonZeroFeaturesDf['abs_Coefficient'] = nonZeroFeaturesDf['Coefficient'].abs()

nonZeroFeaturesDf = nonZeroFeaturesDf.sort_values(by='abs_Coefficient', ascending=False)

plt.figure(figsize=(12, 8))
plt.barh(nonZeroFeaturesDf['Feature'], nonZeroFeaturesDf['abs_Coefficient'], color='skyblue')
plt.xlabel('Absolute Coefficient Value')
plt.ylabel('Feature')
plt.title('Feature Importance (Elastic Net)')
plt.gca().invert_yaxis()
plt.show()
```

```python
selectedX = resampledX[nonZeroFeatures]
selectedX

selected_features = selectedX.columns

```

```python
from sklearn.ensemble import StackingClassifier, RandomForestClassifier, AdaBoostClassifier
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression, RidgeClassifier
from sklearn.calibration import CalibratedClassifierCV
from sklearn.ensemble import GradientBoostingClassifier
import xgboost as xgb

linearSVC = LinearSVC(random_state=42, dual=False, max_iter=2000)
calibratedSVC = CalibratedClassifierCV(linearSVC, method='sigmoid', cv=10)
# randomForest = RandomForestClassifier(n_estimators=300, random_state=42, n_jobs=-1)
xgbc = xgb.XGBClassifier(n_estimators=300, random_state=42, n_jobs=-1)

baseClassifiers = [
    ('svc', calibratedSVC),
    ('xgbc', xgbc)
]

# finalEstimator = xgb.XGBClassifier(n_estimators = 300, random_state = 42, n_jobs=-1)

finalEstimator = LogisticRegression(max_iter=1000, random_state = 42)

# finalEstimator = RidgeClassifier(alpha = 0.15)

stackingClassifier = StackingClassifier(estimators = baseClassifiers, final_estimator = finalEstimator, cv=10, n_jobs=-1)

```

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import confusion_matrix, accuracy_score

kFold = KFold(n_splits = 10, shuffle = True, random_state = 42)
print('We got here!')

confusionMatrices = []
cvScores = []

downloadsPath2 = os.path.expanduser('~/Downloads')

for foldIndex, (trainIndex, testIndex) in enumerate(kFold.split(selectedX)):
    XTrain, XTest = selectedX.iloc[trainIndex], selectedX.iloc[testIndex]
    YTrain, YTest = resampledY.iloc[trainIndex], resampledY.iloc[testIndex]
    
    stackingClassifier.fit(XTrain, YTrain)
    YPred = stackingClassifier.predict(XTest)
    
    cm = confusion_matrix(YTest, YPred)
    confusionMatrices.append(cm)
    
    accuracy = accuracy_score(YTest, YPred)
    cvScores.append(accuracy)
    
    plt.figure(figsize=(10, 7))
    sns.heatmap(cm, annot=True, fmt='g', cmap='Blues', xticklabels=['Poor', 'Standard', 'Good'], yticklabels=['Poor', 'Standard', 'Good'])
    plt.xlabel('Predicted labels')
    plt.ylabel('True labels')
    plt.title(f'Confusion Matrix for Fold {foldIndex+1}\nAccuracy: {accuracy:.4f}')
    pdfPath = os.path.join(downloadsPath2, f'confusion_matrix_fold_{foldIndex+1}.pdf')
    plt.savefig(pdfPath, format='pdf', bbox_inches='tight')
    plt.close()


print(f"Cross-validation scores: {cvScores}")
print(f"Median cross-validation score: {np.median(cvScores)}")
print(f"Mean cross-validation score: {np.mean(cvScores)}")
```
